```python
#create functionality to: 

from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd
from selenium import webdriver
```

```python
flags_url = 'http://flagpedia.net/index'
response = requests.get(flags_url)

# html = response.html
soup = BeautifulSoup(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
soup.find('td', class_='td-country').find('a', href=True)['href']
```

```python
countries = []
for td in soup.find_all(class_='td-country'):
    td_str = str(td)
    td_short = td_str[33:]
#     print(td_short)
    link_end = td_short.find('"')
    country = td_short[:link_end]
#     print('http://flagpedia.net' + href)
    countries.append(country)
countries
```

```python
flag_routes = []
for td in soup.find_all(class_='td-country'):
    td_str = str(td)
    td_short = td_str[32:]
#     print(td_short)
    link_end = td_short.find('"')
    href = td_short[:link_end]
#     print('http://flagpedia.net' + href)
    flag_routes.append('http://flagpedia.net' + href)
flag_routes
```

```python
flag_images = []
for flag in flag_routes:
    response = requests.get(flag)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for id_flag_detail in range(0,1):
        id_flag_detail= soup.find(id='flag-detail')
        id_str =str(id_flag_detail)
        id_short = id_str.find('flags')
        id_start = id_str[id_short:]
    #     print(id_start)
        link_end = id_start.find('"')
    #     print(link_end)
        src = id_start[:link_end]
#         print(src)
        flag_images.append(src)
flag_images
```

```python
flags_dataframe= pd.DataFrame({
    'country':countries,
    'flag_source':flag_routes,
    'flag_image':flag_images
    })
flags_dataframe
```

```python
# create resource 
flags_dataframe.to_csv('flag_images.csv')
```
